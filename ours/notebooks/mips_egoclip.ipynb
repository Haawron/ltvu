```python
p_egoclip = Path('../data/egoclip/egoclip.csv')
# Matches one or more tabs not preceded by double quote
df = pd.read_csv(p_egoclip, sep=r'(?<!")\t+')
df.head()
```

```python
video_uids = set()
with p_egoclip.open() as f:
    while True:
        line = f.readline()
        if not line:
            break
        video_uid = line.split('\t')[0]
        video_uids.add(video_uid)
with p_egoclip.with_name('egoclip_video_uids.txt').open('w') as f:
    f.write('\n'.join(video_uids))
list(video_uids)[:10], len(video_uids)
```

```python
overlapping_video_uids = set(df_nlq['video_uid'].drop_duplicates()) & set(df['video_uid'].drop_duplicates())
len(overlapping_video_uids)
```

```python
clip_uids = df_nlq[df_nlq['video_uid'].isin(overlapping_video_uids)]['clip_uid'].drop_duplicates().reset_index(drop=True)
result = {'clips': []}
for clip_uid in clip_uids:
    df_nlq_clip = df_nlq[df_nlq['clip_uid'] == clip_uid]
    video_uid = df_nlq_clip['video_uid'].values[0]
    clip_s, clip_e = df_nlq_clip[['video_start_sec', 'video_end_sec']].values[0]
    df_clip_narrs = df[
        (df['video_uid'] == video_uid) &
        (clip_s <= df['clip_start']) & (df['clip_end'] <= clip_e)
    ].reset_index(drop=True)
    if len(df_clip_narrs['narration_time']) < 10:
        print(df[df['video_uid'] == video_uid])
        print()
        print(df_clip_narrs)
        print()
        print(df_nlq_clip)
        continue
    clip_instance = {
        'video_uid': video_uid,
        'clip_uid': clip_uid,
        'video_start_sec': clip_s,
        'video_end_sec': clip_e,
        'narrations': {
            'time': df_clip_narrs['narration_time'].tolist(),
            'start': df_clip_narrs['clip_start'].tolist(),
            'end': df_clip_narrs['clip_end'].tolist(),
            'text': df_clip_narrs['clip_text'].tolist(),
        },
        'queries': {
            'start': df_nlq_clip['q_clip_start_sec'].tolist(),
            'end': df_nlq_clip['q_clip_end_sec'].tolist(),
            'query': df_nlq_clip['query'].tolist(),
        },
    }
    result['clips'].append(clip_instance)
```

```python
from pathlib import Path
import numpy as np
import json
import pandas as pd
from collections import defaultdict


p_human_nlq_problem_dir = Path('../data/Ego4D-processed/human_nlq')
with open('../data/Ego4D-processed/captions/VideoRecap/caption_2s/val/VideoRecap_val_gathered.json') as f:
    data = json.load(f)
    data['clips'] = [clip for clip in data['clips'] if (p_human_nlq_problem_dir / clip['clip_uid']).with_suffix('.txt').exists()]

gts = defaultdict(list)
for clip_info in data['clips']:
    clip_uid = clip_info['clip_uid']
    p_txt = p_human_nlq_problem_dir / f'{clip_uid}.txt'
    with p_txt.open() as f:
        lines = f.read().strip().split('\n')
    lines = lines[lines.index('####### Queries #######')+1:]
    lines = [line.split('|')[0].strip() for line in lines]
    df = pd.DataFrame(clip_info['queries'])
    for q in lines:
        found = df[['start', 'end']][df['query'].str.contains(q)].reset_index(drop=True)
        s, e = found.iloc[0].values
        gts[clip_uid].append((s, e, q))
gts = dict(gts)
gts[list(gts.keys())[0]]
```
